```python
class Solution:
    from itertools import product

    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []

        graph = {
            '2' : 'abc',
            '3' : 'def',
            '4' : 'ghi',
            '5' : 'jkl',
            '6' : 'mno',
            '7' : 'pqrs',
            '8' : 'tuv',
            '9' : 'wxyz'
        }

        value = [i for i in digits]
        string = [graph[s] for s in value]

        return [''.join(c) for c in product(*string)]
```

```python
def count_ways_to_express(n):
    # 재귀적으로 백트래킹을 수행하는 함수
    def backtrack(current_sum, remaining_numbers):
        nonlocal count

        # 기저 조건: 현재의 합이 목표 값에 도달하면 count 증가
        if current_sum == n:
            count += 1
            return

        # 현재의 합이 목표 값을 초과하거나 남은 숫자가 없으면 종료
        if current_sum > n or not remaining_numbers:
            return

        # 1, 2, 3 각각을 선택하며 재귀 호출
        for i in range(len(remaining_numbers)):
            next_number = remaining_numbers[i]
            backtrack(current_sum + next_number, remaining_numbers[i:])

    count = 0  # 방법의 수를 저장할 변수

    # 백트래킹 시작
    backtrack(0, [1, 2, 3])

    return count

# 테스트 케이스 입력 및 결과 출력
T = int(input("테스트 케이스의 개수를 입력하세요: "))
for _ in range(T):
    n = int(input())
    result = count_ways_to_express(n)
    print(result)
```

```python
from collections import deque

def subsets_bfs(nums):
    result = [[]]
    queue = deque()

    for num in nums:
        queue.append([num])  # 각 숫자를 리스트로 감싸서 큐에 추가하여 시작

    while queue:
        print(queue)
        current_subset = queue.popleft()
        current_element = current_subset[-1]
        print(queue, current_subset, current_element, '시작')

        # 중복 체크를 통해 불필요한 반복을 방지
        if current_subset not in result:
            result.append(current_subset)

        for num in nums:
            if num > current_element:
                new_subset = current_subset + [num]
                print(new_subset, current_subset, [num])
                queue.append(new_subset)
    
    return result

# 테스트
nums = [1, 2, 3]
output = subsets_bfs(nums)
print(output)
```

```python
from collections import deque

def count_ways_to_express_bfs(n):
    queue = deque([n])  # 큐에 초기값 추가
    ways_count = 0  # 방법의 수를 저장할 변수

    while queue:
        current_number = queue.popleft()

        # 현재 숫자가 0이면 하나의 방법을 찾은 것이므로 count 증가
        if current_number == 0:
            ways_count += 1
        else:
            # 1, 2, 3 각각을 뺀 결과를 큐에 추가
            for step in [1, 2, 3]:
                if current_number - step >= 0:
                    queue.append(current_number - step)

    return ways_count

# 예시 테스트
n = 7
result = count_ways_to_express_bfs(n)
print(result)
```

```python
from collections import deque

graph = [[] for _ in range(numCourses+1)]
indegree = [0]*(numCourses+1)

for i in prerequisites:
    u, v = i
    graph[u].append(v)
    indegree[v] += 1
    
print(graph, indegree)

def topology_sort():
    result = []
    q = deque()
    
    for i in range(1, numCourses+1):
        if indegree[i] == 0:
            q.append(i)
            
    while q:
        current = q.popleft()
        result.append(current)
        
        for i in graph[current]:
            # current로 부터 나가는 간선 제거
            indegree[i] -= 1
            if indegree[i] == 0:
                q.append(i)
    return result
```
